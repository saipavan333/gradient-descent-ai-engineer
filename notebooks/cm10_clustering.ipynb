## k-means: inertia always falls, silhouette peaks at the true k

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.8, random_state=0)
for k in [2, 3, 4]:
    km = KMeans(n_clusters=k, n_init=10, random_state=0).fit(X)
    print(f"k={k}:  inertia {km.inertia_:7.1f}   silhouette {silhouette_score(X, km.labels_):.3f}")
```
