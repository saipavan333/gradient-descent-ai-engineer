## Plan-then-execute: decompose the goal, then work the plan
For complex goals, the agent first drafts an ordered plan, then executes each step (each step would itself use tools / ReAct). Planning keeps the overall goal in view instead of drifting step to step.

```python
goal = "Write a short blog post about coffee"

# PLAN: the model decomposes the goal into ordered subtasks
plan = ["research coffee history", "outline the post", "write the draft", "proofread & polish"]

print("GOAL:", goal)
for i, step in enumerate(plan, 1):
    print(f"  plan step {i}: {step}")

# EXECUTE: work through the plan in order (each step would itself use tools / ReAct)
print("executing:")
for step in plan:
    print(f"  [done] {step}")
print("post complete.")
```
