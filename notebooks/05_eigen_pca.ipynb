# Lesson 1.5 — Eigenvectors, eigenvalues & PCA

```python
import numpy as np
A = np.array([[3.,1.],[1.,3.]])
vals, vecs = np.linalg.eig(A)
print('eigenvalues:', vals.round(2))           # 4 and 2
print('eigenvectors (columns):\n', vecs.round(2))
```

## PCA by hand: compress height & weight to one number

```python
H = np.array([150,160,170,180,190.]); W = np.array([50,58,65,73,80.])
X = np.column_stack([H,W]); Xc = X - X.mean(0)
cov = np.cov(Xc.T)
vals, vecs = np.linalg.eigh(cov); order = np.argsort(vals)[::-1]
vals, vecs = vals[order], vecs[:,order]
print('explained variance ratio:', (vals/vals.sum()).round(4))
pc1 = vecs[:,0]
if pc1[0] < 0: pc1 = -pc1   # eigenvectors are sign-arbitrary; pin sign to match the lesson
print('PC1 direction:', pc1.round(2))
print('each person on PC1:', (Xc @ pc1).round(1))
```

## Same thing with scikit-learn (optional)

```python
try:
    from sklearn.decomposition import PCA
    p = PCA(n_components=1).fit(X)
    print('sklearn explained variance ratio:', p.explained_variance_ratio_.round(4))
except ImportError:
    print('scikit-learn not installed — skip')
```
