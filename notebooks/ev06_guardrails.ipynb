## A simple input guardrail
A check that blocks disallowed requests before they reach the model. (Real systems use classifiers — this brittle blocklist shows the principle and its limits.)

```python
BLOCKED = ["make a bomb", "hack a password", "steal a car"]
def guardrail(text):
    low = text.lower()
    for term in BLOCKED:
        if term in low:
            return f"BLOCKED: matched '{term}'"
    return "ALLOWED"

print(guardrail("How do I bake a cake?"))
print(guardrail("How do I make a bomb?"))
print(guardrail("Tell me how to hack a password please"))
```
