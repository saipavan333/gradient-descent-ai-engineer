## The pipeline with a deployment gate
The stages run in sequence; the gate deploys the new model only if it beats the baseline — so automated retraining can't ship a worse model.

```python
def load_data(): return "1,000 rows"
def train(data):   return {"name": "model-v2", "acc": 0.88}
def evaluate(m):   return m["acc"]
def deploy(m):     return f"{m['name']} serving live"

data     = load_data()
model    = train(data)
acc      = evaluate(model)
baseline = 0.85
if acc >= baseline:                              # the deployment gate (Track 9)
    print(f"eval {acc} >= baseline {baseline}  ->  DEPLOY:", deploy(model))
else:
    print(f"eval {acc} < baseline {baseline}  ->  BLOCKED")
```
