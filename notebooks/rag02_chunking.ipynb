## Fixed-size chunking with overlap

```python
text = "Machine learning learns patterns from data. Models improve with more examples given to them."
def chunk(text, size, overlap):
    chunks, i = [], 0
    while i < len(text):
        chunks.append(text[i:i+size]); i += size - overlap
    return chunks
for j, c in enumerate(chunk(text, size=40, overlap=10)):
    print(f"chunk {j}: {c!r}")
```
