## A bigram language model by counting

```python
from collections import defaultdict, Counter
corpus = "the cat sat on the mat the cat ran".split()
bigrams = defaultdict(Counter)
for a, b in zip(corpus, corpus[1:]):
    bigrams[a][b] += 1
ctx = "the"; counts = bigrams[ctx]; total = sum(counts.values())
print(f"P(next | '{ctx}'):", {w: round(c/total,2) for w,c in counts.items()})
```
