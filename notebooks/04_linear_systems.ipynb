# Lesson 1.4 — Linear systems, span, rank & the inverse
NumPy only.

```python
import numpy as np
A = np.array([[1,1],[2,4]]); b = np.array([5,14])   # items=5, dollars=14
print('solution (muffins, lattes):', np.linalg.solve(A, b))
print('determinant:', round(np.linalg.det(A),3), ' rank:', np.linalg.matrix_rank(A))
```

## A singular (rank-deficient) system fails

```python
A_bad = np.array([[1,2],[2,4]])   # row2 = 2*row1
print('rank:', np.linalg.matrix_rank(A_bad), ' det:', round(np.linalg.det(A_bad),3))
try:
    np.linalg.solve(A_bad, np.array([3,6]))
except np.linalg.LinAlgError as e:
    print('solve failed:', e)
```

## Inverse undoes a transformation

```python
A = np.array([[2.,1.],[1.,3.]]); Ainv = np.linalg.inv(A)
print('A @ A_inv =\n', np.round(A @ Ainv, 6))   # identity
```
