## What a cache hit rate is worth
Savings scale directly with the hit rate: a 60% hit rate means only 40% of requests reach the expensive model.

```python
requests, hit_rate, cost_miss = 1_000_000, 0.60, 0.002
without_cache = requests * cost_miss
with_cache    = requests * (1 - hit_rate) * cost_miss
print(f"without cache        : ${without_cache:,.0f}")
print(f"with {hit_rate:.0%} hit rate    : ${with_cache:,.0f}")
print(f"saved                : ${without_cache - with_cache:,.0f} ({hit_rate:.0%})")
```
