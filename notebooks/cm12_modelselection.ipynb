## Cross-validation gives a stable score; grid search tunes safely (no leakage)

```python
from sklearn.datasets import load_breast_cancer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, GridSearchCV

X, y = load_breast_cancer(return_X_y=True)
pipe = make_pipeline(StandardScaler(), LogisticRegression(max_iter=5000))
scores = cross_val_score(pipe, X, y, cv=5)
print("CV scores:", scores.round(3))
print(f"mean {scores.mean():.3f}  std {scores.std():.3f}")

grid = GridSearchCV(pipe, {"logisticregression__C": [0.01, 0.1, 1, 10]}, cv=5).fit(X, y)
print("best:", grid.best_params_, " CV score:", round(grid.best_score_, 3))
```
