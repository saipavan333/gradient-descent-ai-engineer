## Computing retrieval metrics (recall@k, precision@k, MRR)

```python
retrieved = ["d3", "d7", "d1", "d9", "d2"]   # ranked results
relevant  = {"d1", "d7"}                      # ground-truth relevant
k = 3; topk = retrieved[:k]
recall    = len([d for d in topk if d in relevant]) / len(relevant)
precision = len([d for d in topk if d in relevant]) / k
first = next((i + 1 for i, d in enumerate(retrieved) if d in relevant), 0)
mrr = 1 / first if first else 0.0
print(f"Recall@{k}: {recall:.2f}   Precision@{k}: {precision:.2f}   MRR: {mrr:.2f}")
```
