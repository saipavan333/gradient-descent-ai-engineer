## Reading an A/B test result
Compare conversion rates between control and treatment. The raw lift is the start of the analysis — you then check statistical significance.

```python
control_n, control_conv = 10_000, 500
treat_n,   treat_conv   = 10_000, 560
control_rate = control_conv / control_n
treat_rate   = treat_conv / treat_n
lift = (treat_rate - control_rate) / control_rate
print(f"control   : {control_rate:.1%}")
print(f"treatment : {treat_rate:.1%}")
print(f"lift      : {lift:+.1%}")
```
