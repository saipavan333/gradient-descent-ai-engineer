# Lesson 1.1 — Vectors (runnable companion)
Run each cell top to bottom. Only **NumPy** is required; the PyTorch cell is optional.

```python
import numpy as np

# A house as a feature vector: [size_sqft, bedrooms, age_years, price_k]
house_A = np.array([1800, 3, 12, 350])
house_B = np.array([ 950, 2, 40, 210])
print("house_A =", house_A, " dimension =", house_A.shape[0])
```

## Norm (length) of a vector

```python
v = np.array([3, 2])
print("||v|| =", np.linalg.norm(v), " (= sqrt(13))")
```

## Distance between two data points = norm of their difference

```python
diff = house_A - house_B
print("difference:", diff)
print("distance  :", np.linalg.norm(diff))
# Notice: size and price dominate because they live on much larger scales.
```

## Add and scale

```python
u = np.array([2, 1]); w = np.array([1, 2])
print("u + w      =", u + w)        # element-wise, tip-to-tail
print("1.5 * u    =", 1.5 * u)      # stretch
print("-u         =", -u)           # flip
```

## Three different norms (L1, L2, Linf) on the same vector

```python
x = np.array([3, -4])
print("L1   :", np.linalg.norm(x, 1))      # 7
print("L2   :", np.linalg.norm(x))         # 5
print("Linf :", np.linalg.norm(x, np.inf)) # 4
```

## The mean (average) vector of a dataset

```python
X = np.array([house_A, house_B, [2400,4,5,560]])
print("average house:", X.mean(axis=0))   # axis=0 collapses rows -> per-feature mean
```

## Optional: the same in PyTorch (tensors are vectors with GPU + autograd superpowers)
