## The model emits a structured tool call; YOUR code runs it
The model decides *which* tool and *what arguments*; the system executes the real function and feeds the result back for the final answer.

```python
def get_weather(city):                 # the real tool — YOUR code
    db = {"Paris": "18C cloudy", "Tokyo": "25C sunny"}
    return db.get(city, "unknown")

tools = {"get_weather": get_weather}   # registry of available tools

def llm_decide(question):              # the model emits a structured call (simulated)
    return {"tool": "get_weather", "arguments": {"city": "Tokyo"}}

question = "What's the weather in Tokyo?"
call = llm_decide(question)                              # 1) model decides
result = tools[call["tool"]](**call["arguments"])        # 2) YOUR code executes
print("model emitted tool call:", call)
print("tool result:", result)
print("final answer: It is", result, "in Tokyo.")       # 3) model answers from result
```
