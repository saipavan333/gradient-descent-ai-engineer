# Lesson 1.8 — Gradient descent

```python
# Minimize f(w)=(w-3)^2+1 by hand
w=0.0; lr=0.1
for _ in range(25): w -= lr*2*(w-3)
print('learned w:', round(w,3))   # ~3
```

## Fit a line y=2x with NumPy gradient descent + track the loss

```python
import numpy as np
x=np.array([1.,2,3,4]); y=np.array([2.,4,6,8]); w=0.0; lr=0.01; losses=[]
for _ in range(200):
    pred=w*x; loss=((pred-y)**2).mean(); losses.append(loss)
    grad=(2*x*(pred-y)).mean(); w-=lr*grad
print('learned w:', round(w,3), ' final loss:', round(losses[-1],6))
```

```python
try:
    import matplotlib.pyplot as plt
    plt.plot(losses); plt.xlabel('step'); plt.ylabel('loss'); plt.title('loss going down'); plt.show()
except Exception as e:
    print('plot skipped:', e)
```

## The canonical PyTorch training loop (optional)

```python
try:
    import torch
    x=torch.tensor([1.,2,3,4]); y=torch.tensor([2.,4,6,8]); w=torch.tensor(0.,requires_grad=True)
    opt=torch.optim.SGD([w],lr=0.01)
    for _ in range(200):
        loss=((w*x-y)**2).mean(); opt.zero_grad(); loss.backward(); opt.step()
    print('torch learned w:', round(w.item(),3))
except ImportError:
    print('PyTorch not installed — skip')
```
