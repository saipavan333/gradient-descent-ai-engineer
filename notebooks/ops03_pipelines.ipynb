## One feature definition, no train/serve skew
The same `features` function is used at training and serving time, so the same raw input produces identical features — guaranteed.

```python
import math
def features(row):                       # ONE definition, used for train AND serve
    return {"amount_log": round(math.log1p(row["amount"]), 3),
            "is_weekend": row["day"] in ("Sat", "Sun")}

train_row = {"amount": 100, "day": "Sat"}
serve_row = {"amount": 100, "day": "Sat"}
print("train-time:", features(train_row))
print("serve-time:", features(serve_row))
print("identical :", features(train_row) == features(serve_row))
```
