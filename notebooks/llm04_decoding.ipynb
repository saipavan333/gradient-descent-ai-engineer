## Greedy, temperature, top-k, top-p on one distribution

```python
import numpy as np
def softmax(z, T=1.0):
    z = z / T; e = np.exp(z - z.max()); return e / e.sum()
logits = np.array([3.0, 2.0, 1.0, 0.5, 0.2, -1.0])
print("greedy pick:    token", logits.argmax())
print("T = 1.0:", softmax(logits).round(3))
print("T = 0.5 (sharper):", softmax(logits, 0.5).round(3))
print("T = 1.5 (flatter):", softmax(logits, 1.5).round(3))
p = softmax(logits)
idx = p.argsort()[::-1][:3]; m = np.zeros_like(p); m[idx] = p[idx]
print("top-k = 3:", (m / m.sum()).round(3))
order = p.argsort()[::-1]; cum = np.cumsum(p[order]); keep = order[:np.searchsorted(cum, 0.9) + 1]
m2 = np.zeros_like(p); m2[keep] = p[keep]
print("top-p = 0.9:", (m2 / m2.sum()).round(3))
```
