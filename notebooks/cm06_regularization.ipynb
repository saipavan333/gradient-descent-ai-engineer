## Ridge shrinks all weights; Lasso zeros the useless ones

```python
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler

rng = np.random.default_rng(0)
X = rng.normal(0, 1, (100, 8))
true = np.array([3.0, -2.0, 1.5, 0, 0, 0, 0, 0])   # only first 3 matter
y = X @ true + rng.normal(0, 0.5, 100)
Xs = StandardScaler().fit_transform(X)

print("true ", true)
print("OLS  ", np.round(LinearRegression().fit(Xs, y).coef_, 2))
print("Ridge", np.round(Ridge(alpha=10).fit(Xs, y).coef_, 2))
lasso = Lasso(alpha=0.3).fit(Xs, y).coef_
print("Lasso", np.round(lasso, 2), " zeros:", int((np.abs(lasso) < 1e-8).sum()))
```
