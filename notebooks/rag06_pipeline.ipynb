## A minimal end-to-end RAG pipeline (retrieve + cite + prompt)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
docs = {"[1]": "A refund is processed within 5 business days.",
        "[2]": "You can request a refund within 30 days of purchase.",
        "[3]": "Our stores are open 9am to 6pm on weekdays."}
ids, texts = list(docs), list(docs.values())
vec = TfidfVectorizer().fit(texts)
def retrieve(query, k=2):
    sims = cosine_similarity(vec.transform([query]), vec.transform(texts))[0]
    return [(ids[i], texts[i]) for i in sims.argsort()[::-1][:k]]
query = "how long does a refund take?"
context = retrieve(query)
prompt = ("Answer with a citation [n] using only the context.\n"
          + "".join(f"{cid} {t}\n" for cid, t in context)
          + f"Question: {query}\nAnswer:")
print(prompt)
```
