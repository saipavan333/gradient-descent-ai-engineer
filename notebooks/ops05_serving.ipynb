## The same model, served two ways
The model logic doesn't change between serving patterns — only how inputs arrive. Online answers one request; batch scores many at once.

```python
def predict(x): return x * 2                          # the packaged 'model'

# online: one request at a time, answered immediately
print("online (1 request) :", predict(5))

# batch: score many inputs at once, on a schedule
batch = [1, 2, 3, 4, 5]
print("batch (many at once):", [predict(x) for x in batch])
```
