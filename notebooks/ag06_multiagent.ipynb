## Supervisor/worker: a coordinator delegates to specialists
The supervisor decomposes the goal and routes work to focused workers (a researcher, a writer), passing one's output to the next. Each worker would be its own LLM with a specialized prompt and tools.

```python
def researcher(topic):  # worker 1: focused on gathering info
    return f"3 facts about {topic}"

def writer(notes):      # worker 2: focused on composing prose
    return f"a paragraph based on [{notes}]"

def supervisor(goal):   # coordinator: delegates and assembles
    print(f"supervisor: goal = '{goal}'")
    notes = researcher(goal); print(f"  -> researcher returns: {notes}")
    draft = writer(notes);   print(f"  -> writer returns: {draft}")
    return draft

final = supervisor("electric cars")
print("final output:", final)
```
