## A minimal RAG pipeline
Retrieve the relevant chunk, then ground the answer in it. (TF-IDF stands in for neural embeddings; the pattern is identical.)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
docs = {"[1]": "Our refund policy allows returns within 30 days.",
        "[2]": "The head office is in Seattle.",
        "[3]": "Reset your password from the login page."}
ids, texts = list(docs), list(docs.values())
vec = TfidfVectorizer().fit(texts)
def retrieve(q, k=1):
    sims = cosine_similarity(vec.transform([q]), vec.transform(texts))[0]
    return [ids[i] for i in sims.argsort()[::-1][:k]]
q = "how many days do I have for a refund?"
ctx = retrieve(q)
print("query    :", q)
print("retrieved:", ctx[0], "->", docs[ctx[0]])
```
