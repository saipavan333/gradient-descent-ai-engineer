## Long-term memory: store facts, recall them by cue
The context window is short-term memory; an external store gives long-term recall across steps. Here `recall` matches on a keyword — in production this is embedding similarity (Track 7).

```python
long_term = []
def remember(fact): long_term.append(fact)
def recall(keyword):  # retrieve facts matching the cue (embedding similarity in production)
    return [m for m in long_term if keyword.lower() in m.lower()]

remember("The user's name is Alex.")
remember("The user prefers Python over Java.")
remember("The user is allergic to peanuts.")

print("stored", len(long_term), "facts")
print("recall('python') ->", recall("python"))
print("recall('peanut') ->", recall("peanut"))
```
