## An agent loop with guardrails
Think -> act -> observe, but every action passes a permission allowlist and a budget first — so a disallowed action can't run.

```python
def calculator(expr): return eval(expr)        # a tool
tools = {"calculator": calculator}
ALLOWED = {"calculator"}                        # guardrail: permission allowlist
budget, used = 2, 0

def agent(goal):
    global used
    print(f"goal: {goal}")
    plan = [("calculator", "12*8"), ("delete_files", "/"), ("calculator", "100-58")]
    for tool, arg in plan:                       # think -> act -> observe, with guardrails
        if tool not in ALLOWED:
            print(f"  BLOCKED: '{tool}' not permitted"); continue
        if used >= budget:
            print(f"  BLOCKED: budget {budget} exceeded"); continue
        used += 1
        print(f"  Thought: use {tool}  ->  Action: {tool}({arg})  ->  Observation: {tools[tool](arg)}")
    return "done"
agent("compute some values, safely")
```
