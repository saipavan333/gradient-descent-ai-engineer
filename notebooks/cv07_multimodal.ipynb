## Zero-shot matching with cosine similarity (CLIP idea)


```python
import numpy as np
def cos(a, b):
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))

img_emb = np.array([0.9, 0.1, 0.8, 0.2])     # the image's vector (from the image encoder)
captions = {                                  # candidate labels, written as text
    "a photo of a cat": np.array([0.85, 0.15, 0.75, 0.10]),
    "a photo of a dog": np.array([0.20, 0.90, 0.15, 0.80]),
    "a red bicycle":    np.array([0.10, 0.20, 0.10, 0.95]),
}
scores = {c: cos(img_emb, v) for c, v in captions.items()}
for c, s in scores.items():
    print(f"  {c:22s} cos = {s:.3f}")
print("best match:", repr(max(scores, key=scores.get)))
```

## Explore: change the label set and reclassify instantly (no retraining)


```python
# Zero-shot: invent new categories on the spot, just by writing text vectors.
new_labels = {
    "an animal":   np.array([0.80, 0.30, 0.70, 0.25]),
    "a vehicle":   np.array([0.10, 0.25, 0.10, 0.92]),
}
for c, v in new_labels.items():
    print(f"  {c:12s} cos = {cos(img_emb, v):.3f}")
print("-> the same image vector is reused; only the text labels changed.")
```
