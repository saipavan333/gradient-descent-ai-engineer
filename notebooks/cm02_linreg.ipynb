## Fit a line two ways: normal equations (NumPy) and scikit-learn

```python
import numpy as np
rng = np.random.default_rng(0)
n = 200
X = rng.uniform(0, 10, (n, 1))
y = 2.5 * X[:, 0] + 4 + rng.normal(0, 1.5, n)   # true: slope 2.5, intercept 4

X_b = np.c_[np.ones(n), X]
beta = np.linalg.solve(X_b.T @ X_b, X_b.T @ y)
print("from scratch  -> intercept, slope:", beta.round(3))
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X, y)
print("scikit-learn  -> intercept, slope:", round(model.intercept_,3), round(model.coef_[0],3))
mse = ((model.predict(X) - y)**2).mean()
print("MSE:", round(mse, 3))
```

Both routes give the same line because scikit-learn solves the same normal equations. MSE ~ the noise variance (1.5**2=2.25) is the best achievable here.
