# Lesson 2.5 — Cleaning, encoding, and avoiding leakage

```python
import pandas as pd, numpy as np
df = pd.DataFrame({"age":[34,np.nan,41,np.nan,52], "color":["red","green","blue","red","green"]})
df["age"] = df["age"].fillna(df["age"].mean())     # impute
print(df)
print("\none-hot:\n", pd.get_dummies(df["color"], prefix="color").astype(int))
```

## Fit the scaler on TRAIN only (no leakage)

```python
import numpy as np
try:
    from sklearn.model_selection import train_test_split
    from sklearn.preprocessing import StandardScaler
    X = np.array([[1800.],[950.],[2400.],[1500.],[1200.],[3000.]]); y=np.array([350,210,560,300,250,700])
    Xtr,Xte,ytr,yte = train_test_split(X,y,test_size=0.33,random_state=0)
    sc = StandardScaler().fit(Xtr)                 # fit on TRAIN only
    print("train mean:", sc.mean_.round(1), " std:", sc.scale_.round(1))
    print("test scaled:", sc.transform(Xte).ravel().round(2))
except ImportError:
    print("scikit-learn not installed — skip")
```
