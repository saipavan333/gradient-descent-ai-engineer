## Fit a line with gradient descent
A classic ML coding question: fit y = w*x (true w = 3) by repeatedly stepping w opposite the gradient of the mean-squared error. It converges to the true slope.

```python
xs = [1, 2, 3, 4]
ys = [3, 6, 9, 12]                 # true relationship: y = 3x
w, lr = 0.0, 0.01                  # start at 0, learning rate 0.01
for step in range(1000):
    # gradient of mean squared error wrt w:  mean of 2*(w*x - y)*x
    grad = sum(2 * (w * x - y) * x for x, y in zip(xs, ys)) / len(xs)
    w -= lr * grad                 # step downhill
print(f"learned w: {w:.3f}  (true 3.0)")
print(f"predict x=5 -> {w*5:.2f}")
```

## Bonus: another from-scratch classic — cosine similarity
Used everywhere in embeddings and retrieval (Track 7). Two vectors are similar when the cosine of the angle between them is near 1.

```python
import math
def cosine(a, b):
    dot = sum(x*y for x, y in zip(a, b))
    na = math.sqrt(sum(x*x for x in a))
    nb = math.sqrt(sum(y*y for y in b))
    return dot / (na * nb)

print("cosine([1,0,1],[1,1,0]) =", round(cosine([1,0,1],[1,1,0]), 4))   # 0.5
print("cosine([1,2,3],[2,4,6]) =", round(cosine([1,2,3],[2,4,6]), 4))   # 1.0 (parallel)
```
