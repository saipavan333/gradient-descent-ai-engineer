## A two-layer MLP forward pass (ReLU hidden, linear output)

```python
import numpy as np
rng = np.random.default_rng(0)
x  = np.array([1.0, 2.0, -1.0])
W1 = rng.normal(0,1,(4,3)); b1 = np.zeros(4)
W2 = rng.normal(0,1,(1,4)); b2 = np.zeros(1)
h = np.maximum(0, W1 @ x + b1)
y = W2 @ h + b2
print("hidden (after ReLU):", h.round(3))
print("output:", y.round(3))
```
