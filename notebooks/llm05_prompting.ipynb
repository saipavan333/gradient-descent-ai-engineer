## Building a few-shot prompt (string assembly)

```python
examples = [("happy", "positive"), ("terrible", "negative"), ("okay", "neutral")]
query = "wonderful"
prompt = "Classify sentiment.\n"
for word, label in examples:
    prompt += f"Text: {word} -> {label}\n"
prompt += f"Text: {query} ->"
print(prompt)
```
