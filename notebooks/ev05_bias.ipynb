## Detecting disparate impact
Compare each group's favorable-outcome rate against the best group's. A ratio below 0.80 (the four-fifths rule) is a red flag.

```python
groups = {"A": (600, 1000), "B": (420, 1000)}   # (approved, total) per group
rates = {g: a / t for g, (a, t) in groups.items()}
best = max(rates.values())
for g, r in rates.items():
    ratio = r / best
    flag = "OK" if ratio >= 0.80 else "DISPARATE IMPACT"
    print(f"group {g}: rate {r:.0%}  ratio {ratio:.2f}  [{flag}]")
```
