## Backprop by hand, checked against a numerical gradient

```python
x, w, v, t = 2.0, 0.5, 1.5, 3.0
z = w*x; a = max(0.0, z); yhat = a*v; L = (yhat - t)**2
dL_dyhat = 2*(yhat - t); dyhat_da = v; da_dz = 1.0 if z>0 else 0.0; dz_dw = x
grad = dL_dyhat*dyhat_da*da_dz*dz_dw
print("loss:", L, " analytic dL/dw:", grad)
eps = 1e-6
def Lf(w): z=w*x; a=max(0.0,z); return (a*v - t)**2
print("numerical dL/dw:", round((Lf(w+eps)-Lf(w-eps))/(2*eps), 5))
```
