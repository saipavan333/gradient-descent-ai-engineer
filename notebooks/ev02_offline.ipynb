## Precision, recall, and F1
From a confusion matrix (TP/FP/FN/TN), the honest metrics follow directly. Precision and recall trade off; F1 balances them.

```python
TP, FP, FN, TN = 80, 30, 20, 870
precision = TP / (TP + FP)        # of what we flagged, how much was right?
recall    = TP / (TP + FN)        # of the real positives, how many did we catch?
f1 = 2 * precision * recall / (precision + recall)   # their harmonic mean
print(f"precision: {precision:.3f}")
print(f"recall   : {recall:.3f}")
print(f"F1       : {f1:.3f}")
```
