## Many trees beat one (5-fold cross-validated accuracy)

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

X, y = load_breast_cancer(return_X_y=True)
for name, model in [("single tree", DecisionTreeClassifier(random_state=0)),
                    ("random forest", RandomForestClassifier(n_estimators=300, random_state=0)),
                    ("grad boosting", GradientBoostingClassifier(random_state=0))]:
    s = cross_val_score(model, X, y, cv=5)
    print(f"{name:14s} {s.mean():.3f} +/- {s.std():.3f}")
```
