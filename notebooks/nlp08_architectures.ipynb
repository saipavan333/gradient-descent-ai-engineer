## The causal mask that makes a decoder (lower-triangular attention)

```python
import numpy as np
softmax = lambda x: np.exp(x-x.max(-1,keepdims=True))/np.exp(x-x.max(-1,keepdims=True)).sum(-1,keepdims=True)
scores = np.array([[2.,1.,.5,1.],[1.,3.,.5,1.],[.5,1.,2.,1.],[1.,1.,1.,3.]])
mask = np.triu(np.ones((4,4)), k=1).astype(bool)
m = scores.copy(); m[mask] = -np.inf
A = softmax(m)
print("causal attention (each token sees only itself + past):\n", A.round(2))
```
