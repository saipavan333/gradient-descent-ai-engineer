## Full self-attention on a sequence: softmax(QKᵀ/√d)·V

```python
import numpy as np
def softmax(x):
    e = np.exp(x - x.max(axis=-1, keepdims=True))
    return e / e.sum(axis=-1, keepdims=True)
X = np.array([[1,0,1,0],[0,1,0,1],[1,1,0,0]], float)
d = X.shape[1]
Q = K = V = X
scores = Q @ K.T / np.sqrt(d)
A = softmax(scores)
output = A @ V
print("attention matrix (rows sum to 1):\n", A.round(2))
print("row sums:", A.sum(axis=1).round(3))
print("output:\n", output.round(2))
```
