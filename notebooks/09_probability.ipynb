# Lesson 1.9 — Probability & Bayes

```python
# Medical test: rare disease, accurate test, surprising posterior
p_sick=0.01; sens=0.99; fpr=0.05
p_pos=sens*p_sick + fpr*(1-p_sick)
print('P(sick | positive):', round(sens*p_sick/p_pos, 3))   # 0.167
```

## Probability emerges from data (Law of Large Numbers)

```python
import numpy as np
rng=np.random.default_rng(0); flips=rng.integers(0,2,size=10000)
for n in [10,100,1000,10000]:
    print(f'after {n:5d} flips, fraction heads = {flips[:n].mean():.3f}')
```

## Naive-Bayes-style spam update

```python
def posterior(prior, like_spam, like_ham):
    num=like_spam*prior
    return num/(num + like_ham*(1-prior))
print('P(spam | "lottery"):', round(posterior(0.3, 0.10, 0.001), 3))   # ~0.977
```
