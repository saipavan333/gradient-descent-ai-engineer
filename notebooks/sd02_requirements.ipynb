## Sizing the fleet from latency and QPS
Given peak QPS and per-request latency, estimate how many replicas you need (one worker at 100ms/request serves ~10 req/s).

```python
import math
peak_qps   = 6_944
latency_ms = 100
per_replica_qps = 1000 / latency_ms     # 1 worker, 100ms/request -> 10 req/s
replicas = math.ceil(peak_qps / per_replica_qps)
print(f"per-replica QPS : {per_replica_qps:.0f}")
print(f"replicas needed : {replicas}")
print(f"with 30% headroom: {math.ceil(replicas * 1.3)}")
```
