## Exact nearest-neighbor search (what ANN approximates)

```python
import numpy as np
rng = np.random.default_rng(0)
db = rng.standard_normal((1000, 8)); db /= np.linalg.norm(db, axis=1, keepdims=True)
query = db[42] + 0.1 * rng.standard_normal(8); query /= np.linalg.norm(query)
sims = db @ query
top5 = sims.argsort()[::-1][:5]
print("exact top-5 indices:", list(map(int, top5)))
print("their similarities :", sims[top5].round(3))
```
