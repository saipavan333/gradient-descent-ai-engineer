## SFT concentrates probability on helpful responses (toy logits)

```python
import numpy as np
def softmax(z):
    e = np.exp(z - z.max()); return e / e.sum()
base_logits = np.array([2.0, 1.9, 1.8])   # helpful, curt, off-topic
sft_logits  = np.array([4.0, 1.5, 1.0])
print("base model :", softmax(base_logits).round(2))
print("after SFT  :", softmax(sft_logits).round(2))
```
