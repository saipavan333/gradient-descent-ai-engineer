## Estimating the load you must serve
A back-of-the-envelope QPS estimate sizes the problem: 10M daily users × ~20 requests, over 86,400 seconds/day, with a 3x peak factor.

```python
DAU = 10_000_000
req_per_user = 20
sec_per_day = 86_400
avg_qps  = DAU * req_per_user / sec_per_day
peak_qps = avg_qps * 3          # peak is ~3x average (busy hours)
print(f"avg QPS : {avg_qps:,.0f}")
print(f"peak QPS: {peak_qps:,.0f}")
```
