## Reranking changes the order

```python
import numpy as np
docs   = ["A", "B", "C", "D", "E"]
dense  = np.array([0.80, 0.78, 0.60, 0.55, 0.40])   # bi-encoder (fast, coarse)
rerank = np.array([0.30, 0.95, 0.20, 0.88, 0.10])   # cross-encoder (slow, precise)
print("retrieval order (bi-encoder):", [docs[i] for i in dense.argsort()[::-1]])
print("rerank order (cross-encoder):", [docs[i] for i in rerank.argsort()[::-1]])
```
