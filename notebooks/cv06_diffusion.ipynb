## One forward-noising step (and undoing it)


```python
import numpy as np
np.random.seed(0)
x0 = np.array([0.8, 0.2, 0.5, 0.9])   # a clean signal (4 pixels)
alpha_bar = 0.64                       # fraction of signal retained at step t
noise = np.random.randn(4)             # standard Gaussian noise
xt = np.sqrt(alpha_bar)*x0 + np.sqrt(1-alpha_bar)*noise   # the forward step
print("clean x0:   ", np.round(x0,3))
print("noise:      ", np.round(noise,3))
print("sqrt(a)=", round(np.sqrt(alpha_bar),3), " sqrt(1-a)=", round(np.sqrt(1-alpha_bar),3))
print("noisy x_t:  ", np.round(xt,3))
pred_noise = noise
x0_recovered = (xt - np.sqrt(1-alpha_bar)*pred_noise)/np.sqrt(alpha_bar)
print("recovered x0:", np.round(x0_recovered,3))
```

## Explore: as alpha_bar shrinks, the image dissolves into noise


```python
# Higher t -> less signal retained -> x_t looks more like pure noise.
for ab in [0.99, 0.64, 0.25, 0.01]:
    xt = np.sqrt(ab)*x0 + np.sqrt(1-ab)*noise
    print(f"alpha_bar={ab:>4}: signal weight {np.sqrt(ab):.2f}, "
          f"noise weight {np.sqrt(1-ab):.2f} -> x_t {np.round(xt,2)}")
# Generation runs this backward: start at pure noise, denoise step by step.
```
