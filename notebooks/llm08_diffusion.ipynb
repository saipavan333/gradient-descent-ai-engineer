## The forward noising schedule: signal vs noise weight

```python
import numpy as np
for t, abar in [(0, 1.0), (250, 0.7), (500, 0.4), (1000, 0.04)]:
    signal_weight = np.sqrt(abar); noise_weight = np.sqrt(1 - abar)
    print(f"t={t:4d}:  signal weight {signal_weight:.2f},  noise weight {noise_weight:.2f}")
```
