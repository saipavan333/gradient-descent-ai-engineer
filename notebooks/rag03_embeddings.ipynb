## Ranking chunks by cosine similarity (TF-IDF stands in for a neural embedder)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
docs = ["You can get a refund within 30 days of purchase.",
        "Our offices are in Seattle, Washington.",
        "Reset your password via the Forgot Password link.",
        "A return requires the original receipt and packaging."]
query = "how do I get a refund for a return?"
vec = TfidfVectorizer().fit(docs + [query])
sims = cosine_similarity(vec.transform([query]), vec.transform(docs))[0]
for r in sims.argsort()[::-1][:3]:
    print(f"sim {sims[r]:.3f}: {docs[r]}")
```
