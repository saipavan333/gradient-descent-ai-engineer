## A tiny RNN's running memory, and why gradients vanish

```python
import numpy as np
seq = [1.0, 0.5, -1.0, 2.0]; Wh, Wx, h = 0.8, 0.5, 0.0
for t, x in enumerate(seq):
    h = np.tanh(Wh*h + Wx*x)
    print(f"step {t+1}: input {x:+.1f}  ->  hidden state {h:+.3f}")
```

```python
for slope in [0.5, 1.5]:
    print(f"slope {slope} over 20 steps -> gradient x {slope**20:.2e}")
```
