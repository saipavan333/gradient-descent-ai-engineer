## An agent is an LLM in a loop: think → act → observe → repeat
The `think` function stands in for the model deciding the next action; the loop acts and observes until the goal is met. A max-steps guard (Lesson 8.7) prevents runaway loops.

```python
def think(n):                          # the "LLM" decides the next action
    return "stop" if n * n > 50 else "increment"

n, steps = 0, 0
while steps < 20:                      # a max-steps guard (Lesson 8.7)
    decision = think(n)                # THINK: reason about what to do
    print(f"step {steps}: n={n}, n^2={n*n}, decision={decision}")
    if decision == "stop":             # goal met -> exit the loop
        break
    n += 1                             # ACT (and OBSERVE the new n)
    steps += 1
print(f"agent stopped at n={n} (n^2={n*n} > 50)")
```
