## Validating a judge against humans
Before trusting an LLM judge, measure its agreement with human ratings on a labeled sample.

```python
human = [5, 4, 2, 5, 3]          # human ratings (1-5)
judge = [5, 4, 3, 5, 3]          # LLM-judge ratings
exact  = sum(h == j for h, j in zip(human, judge))
within1 = sum(abs(h - j) <= 1 for h, j in zip(human, judge))
n = len(human)
print(f"exact agreement   : {exact}/{n} = {exact/n:.0%}")
print(f"within-1 agreement: {within1}/{n} = {within1/n:.0%}")
```
