# Lesson 2.3 — Broadcasting & axes

```python
import numpy as np
X = np.array([[1800,3,12],[950,2,40],[2400,4,5],[1500,3,25]], dtype=float)
col_mean=X.mean(axis=0); col_std=X.std(axis=0)
Z=(X-col_mean)/col_std
print("col means:", col_mean.round(1))
print("standardized col means:", Z.mean(0).round(6), " stds:", Z.std(0).round(6))
```

## keepdims fixes row-wise broadcasting

```python
row_mean = X.mean(axis=1, keepdims=True)
print("row_mean shape:", row_mean.shape)   # (4,1)
print((X-row_mean).shape)                   # (4,3) broadcasts
```
