## Vectorization beats a loop

```python
import time
rng = np.random.default_rng(0); x = rng.normal(50,10,size=1_000_000); mu,sd = x.mean(), x.std()
t=time.perf_counter()
z=np.empty_like(x)
for i in range(x.size): z[i]=(x[i]-mu)/sd
loop=time.perf_counter()-t
t=time.perf_counter(); zv=(x-mu)/sd; vec=time.perf_counter()-t
print(f"loop {loop:.3f}s  vectorized {vec:.4f}s  speedup ~{loop/vec:.0f}x  same={np.allclose(z,zv)}")
```
