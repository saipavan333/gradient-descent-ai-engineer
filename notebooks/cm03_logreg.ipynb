## Classify tumors; scale on train only; read the probability

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0, stratify=y)

scaler = StandardScaler().fit(X_train)
X_train_s = scaler.transform(X_train)
X_test_s  = scaler.transform(X_test)

clf = LogisticRegression(max_iter=5000).fit(X_train_s, y_train)
print("test accuracy:", round(clf.score(X_test_s, y_test), 3))
```

```python
probs = clf.predict_proba(X_test_s[:1])[0]
print("P(malignant), P(benign):", probs.round(3))
print("true label (0=malignant, 1=benign):", int(y_test[0]))
```
