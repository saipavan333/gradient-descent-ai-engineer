## Intersection over Union (IoU): does a predicted box match the truth?


```python
def iou(a, b):                                  # boxes as (x1, y1, x2, y2)
    ix1, iy1 = max(a[0], b[0]), max(a[1], b[1])
    ix2, iy2 = min(a[2], b[2]), min(a[3], b[3])
    inter = max(0, ix2 - ix1) * max(0, iy2 - iy1)              # overlap area
    union = (a[2]-a[0])*(a[3]-a[1]) + (b[2]-b[0])*(b[3]-b[1]) - inter
    return inter / union

gt, pred = (0, 0, 10, 10), (1, 1, 11, 11)
val = iou(gt, pred)
print(f"IoU = {val:.3f}  ->", "correct (>= 0.5)" if val >= 0.5 else "miss (< 0.5)")
```

## Explore: IoU falls off as the box drifts away


```python
for shift in [0, 1, 3, 5, 9]:
    p = (shift, shift, 10+shift, 10+shift)
    print(f"shift {shift:>2}px -> IoU {iou(gt,p):.3f}",
          "(correct)" if iou(gt,p)>=0.5 else "(miss)")
# Perfect overlap = 1.0; no overlap = 0.0; the 0.5 threshold decides correctness.
```
