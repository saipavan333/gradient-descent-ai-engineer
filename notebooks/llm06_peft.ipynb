## How much does LoRA save?

```python
d, r = 4096, 8
full = d * d
lora = 2 * d * r
print(f"full fine-tune of one weight: {full:,} params")
print(f"LoRA (rank {r}):              {lora:,} params")
print(f"reduction: {full // lora}x fewer  ({lora / full * 100:.3f}% of the weights)")
```
