## PCA: how much variance the top components keep

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X, _ = load_breast_cancer(return_X_y=True)
Xs = StandardScaler().fit_transform(X)
evr = PCA().fit(Xs).explained_variance_ratio_
print("variance explained by first 5 PCs:", evr[:5].round(3))
print(f"cumulative - first 2 PCs: {evr[:2].sum():.3f}   first 10 PCs: {evr[:10].sum():.3f}")
```
