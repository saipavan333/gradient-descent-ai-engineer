## One neuron: weighted sum + bias, through an activation

```python
import numpy as np
x = np.array([2.0, -1.0, 3.0]); w = np.array([0.5, -1.0, 0.25]); b = 0.5
z = w @ x + b
print("z =", z)
print("ReLU(z)    =", max(0.0, z))
print("sigmoid(z) =", round(1/(1+np.exp(-z)), 4))
```
