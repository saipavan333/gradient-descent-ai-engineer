## Guardrails gate every action: permission + budget
Each action must pass independent checks before it runs — a permission allowlist and a budget cap. These hold regardless of what the model outputs: defense in depth.

```python
ALLOWED = {"search", "calculator"}   # permission allowlist
BUDGET = 2                           # max actions per task
count = 0

def guarded(tool, arg):
    global count
    if tool not in ALLOWED:
        return f"BLOCKED: '{tool}' not permitted"
    if count >= BUDGET:
        return f"BLOCKED: budget {BUDGET} exceeded"
    count += 1
    return f"OK: ran {tool}({arg}) [{count}/{BUDGET} actions]"

print(guarded("search", "weather"))     # allowed, within budget
print(guarded("delete_files", "/"))     # not on allowlist
print(guarded("calculator", "2+2"))     # allowed, uses last budget
print(guarded("search", "news"))        # budget now exhausted
```
