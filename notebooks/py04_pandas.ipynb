# Lesson 2.4 — pandas Series & DataFrames

```python
import pandas as pd
orders = pd.DataFrame({"customer":["Ann","Bo","Ann","Cy","Bo"],
                       "region":["West","East","West","East","East"],
                       "amount":[120,80,200,50,300]})
print("shape:", orders.shape)
print("\nbig orders:\n", orders[orders["amount"]>100])
print("\nrevenue per region:\n", orders.groupby("region")["amount"].sum())
```

## merge two tables on a key

```python
customers = pd.DataFrame({"customer":["Ann","Bo","Cy"],"tier":["gold","silver","gold"]})
merged = orders.merge(customers, on="customer")
print(merged.groupby("tier")["amount"].sum())
```
