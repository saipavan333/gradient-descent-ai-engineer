## Why LoRA is cheap: the parameter math
Instead of the full d x d weight update, LoRA trains two small low-rank matrices — a tiny fraction of the parameters.

```python
d, r = 1000, 8
full = d * d                      # full fine-tune updates d x d weights
lora = d * r + r * d              # LoRA: two small low-rank matrices (rank r)
print(f"full fine-tune params : {full:,}")
print(f"LoRA (rank {r}) params : {lora:,}")
print(f"trainable: {lora/full:.1%} of full  ({full//lora}x fewer)")
```
