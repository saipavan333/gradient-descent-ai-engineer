## Counting an LLM's parameters (GPT-2 small)

```python
d, L, vocab = 768, 12, 50257
embedding = vocab * d
per_layer = 12 * d**2
total = embedding + L * per_layer
print(f"embedding: {embedding/1e6:.1f}M")
print(f"per layer: {per_layer/1e6:.2f}M  x{L} layers = {L*per_layer/1e6:.1f}M")
print(f"total ~ {total/1e6:.0f}M  parameters")
```
