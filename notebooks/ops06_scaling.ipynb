## Caching cuts the work
A cache stores each input's result; repeated inputs are served from storage instead of re-running the model. Count how many times the model actually runs.

```python
cache = {}
calls = 0
def predict(x):
    global calls
    if x in cache: return cache[x]       # cache hit: skip the model
    calls += 1                            # cache miss: run the (costly) model
    cache[x] = x * x
    return cache[x]

requests = [3, 3, 5, 3, 5, 7]
results = [predict(x) for x in requests]
print("results   :", results)
print("model runs:", calls, "for", len(requests), "requests")
```
