## Interleave Thought / Action / Observation
The model reasons (Thought), calls a tool (Action), and reads the result (Observation), looping until it can answer. Thinking before each action makes tool use deliberate and the trace debuggable.

```python
def search(q):                          # a tool (your code)
    return {"France population": "68 million", "Germany population": "84 million"}[q]

# the model interleaves Thought / Action; your code executes Actions -> Observation
print("Thought: I need France's population.")
print("Action: search('France population')")
print("Observation:", search("France population"))

print("Thought: Now I need Germany's population.")
print("Action: search('Germany population')")
print("Observation:", search("Germany population"))

print("Thought: 68 + 84 = 152 million. I have enough to answer.")
print("Answer: Together they have about 152 million people.")
```
