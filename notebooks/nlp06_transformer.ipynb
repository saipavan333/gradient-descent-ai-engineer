## One full transformer block in NumPy (shape preserved)

```python
import numpy as np
np.random.seed(0)
softmax = lambda x: np.exp(x-x.max(-1,keepdims=True))/np.exp(x-x.max(-1,keepdims=True)).sum(-1,keepdims=True)
layernorm = lambda x: (x-x.mean(-1,keepdims=True))/(x.std(-1,keepdims=True)+1e-5)
X = np.random.randn(3, 8)
A = softmax(X @ X.T / np.sqrt(8)); x1 = layernorm(X + A @ X)
W1, W2 = np.random.randn(8,16), np.random.randn(16,8)
out = layernorm(x1 + np.maximum(0, x1 @ W1) @ W2)
print("input shape:", X.shape, "-> output shape:", out.shape, "(same!)")
print("output mean per token (~0):", out.mean(-1).round(3))
print("output std  per token (~1):", out.std(-1).round(3))
```
