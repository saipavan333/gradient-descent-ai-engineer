## Why the funnel is essential — the cost math
Ranking the whole catalog with the heavy model vs ranking only the candidates the cheap stage produced.

```python
catalog, candidates, rank_cost = 10_000_000, 500, 100
naive     = catalog    * rank_cost     # rank the whole catalog with the heavy model
two_stage = candidates * rank_cost     # rank only the 500 candidates
print(f"rank all {catalog:,}: {naive:,} units")
print(f"rank {candidates} candidates: {two_stage:,} units")
print(f"cheaper by: {naive // two_stage:,}x")
```
