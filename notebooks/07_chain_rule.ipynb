## Backprop by hand on a tiny graph, checked by autograd

```python
# L=(w*x+b)^2 ; chain: dL/dw = 2a * 1 * x
w,x,b=2.,3.,1.; z=w*x; a=z+b; L=a**2
print('dL/dw by hand:', 2*a*1*x)   # 42
try:
    import torch
    W=torch.tensor(2.,requires_grad=True); X=torch.tensor(3.,requires_grad=True); B=torch.tensor(1.,requires_grad=True)
    ((W*X+B)**2).backward()
    print('autograd dL/dw,db,dx:', W.grad.item(), B.grad.item(), X.grad.item())  # 42,14,28
except ImportError:
    print('PyTorch not installed — skip')
```
