## The retrieve-then-generate loop (keyword retrieval + prompt assembly)

```python
corpus = {
  "doc1": "Our return policy allows refunds within 30 days of purchase.",
  "doc2": "The headquarters is located in Seattle, Washington.",
  "doc3": "To reset your password, click Forgot Password on the login page."}
query = "what is the return policy?"
score = lambda text: len(set(query.lower().replace("?","").split()) & set(text.lower().split()))
best_id, best_text = max(corpus.items(), key=lambda kv: score(kv[1]))
print("retrieved:", best_id)
prompt = f"Answer using ONLY this context.\nContext: {best_text}\nQuestion: {query}\nAnswer:"
print(prompt)
```
