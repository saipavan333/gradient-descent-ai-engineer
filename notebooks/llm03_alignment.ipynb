## Reward -> preference (Bradley-Terry / sigmoid)

```python
import numpy as np
sigmoid = lambda x: 1 / (1 + np.exp(-x))
reward_A, reward_B = 2.5, 1.0
print("P(human prefers A):", round(float(sigmoid(reward_A - reward_B)), 3))
```
