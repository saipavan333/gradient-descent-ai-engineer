## Attention for a single query over keys/values

```python
import numpy as np
q = np.array([1.0, 0.0])
K = np.array([[1.0,0.0],[0.0,1.0],[0.9,0.1]])
V = np.array([[10.0,0.0],[0.0,10.0],[5.0,5.0]])
scores = K @ q
weights = np.exp(scores)/np.exp(scores).sum()
output = weights @ V
print("scores (q·each key):", scores.round(2))
print("attention weights:  ", weights.round(3))
print("output:             ", output.round(2))
```
