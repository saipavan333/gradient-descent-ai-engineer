## Inverted dropout preserves the expected activation

```python
import numpy as np
rng = np.random.default_rng(0)
h = rng.normal(0,1,8); p = 0.5
masks = rng.random((20000,8)) > p
dropped = h * masks / (1-p)
print("original h        :", h.round(2))
print("avg over droppings:", dropped.mean(axis=0).round(2))
```
