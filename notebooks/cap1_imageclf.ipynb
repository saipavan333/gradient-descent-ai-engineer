## Train only the head on frozen features
The transfer-learning principle: a frozen backbone provides features; a tiny trained head adapts them to your task. (Real project: a pretrained CNN + a new linear layer.)

```python
import numpy as np
rng = np.random.default_rng(0)
X = rng.standard_normal((200, 4))                 # features from a FROZEN backbone
true_w = np.array([1.5, -2.0, 0.5, 1.0])
y = (X @ true_w + rng.standard_normal(200)*0.1 > 0).astype(int)   # our labels
w = np.zeros(4)                                    # train ONLY a small linear head
for _ in range(3000):
    p = 1 / (1 + np.exp(-(X @ w)))
    w -= 0.5 * (X.T @ (p - y) / len(y))
acc = ((X @ w > 0).astype(int) == y).mean()
print(f"trained head on frozen features -> accuracy: {acc:.2f}")
```
