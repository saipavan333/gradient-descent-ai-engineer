# Lesson 2.7 — An EDA workflow

```python
import pandas as pd, numpy as np
df = pd.DataFrame({"customer_id":[1,2,3,4,5,5],
                   "age":[34,28,np.nan,45,52,52],
                   "income_k":[55,40,62,90,120,120],
                   "channel":["web","web","app","app","web","web"],
                   "spend":[12,8,15,30,45,45]})
print("shape:", df.shape)
print("\nmissing per column:\n", df.isna().sum())
print("\nduplicate rows:", df.duplicated().sum())
print("\nchannel counts:\n", df["channel"].value_counts())
print("\ncorr with spend:\n", df.corr(numeric_only=True)["spend"].round(2))
```
