## The two-stage cost split, for search
Cheap retrieval scores every document; an expensive reranker refines only the top-k.

```python
docs, rerank_k = 10_000_000, 500
retrieve_cost, rerank_cost = 1, 50
stage1 = docs     * retrieve_cost      # cheap: score every document
stage2 = rerank_k * rerank_cost        # expensive: rerank the top-k
print(f"stage 1 retrieve all : {stage1:,}")
print(f"stage 2 rerank top {rerank_k}: {stage2:,}")
print(f"two-stage total: {stage1 + stage2:,}")
print(f"rerank everything: {docs * rerank_cost:,}")
```
