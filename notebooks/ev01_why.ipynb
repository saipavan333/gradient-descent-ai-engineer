## The accuracy paradox
A lazy model that always predicts 'not fraud' scores 99% accuracy on imbalanced data — yet catches zero fraud. The metric measured the wrong thing.

```python
total, fraud = 1000, 10          # 1% of transactions are fraud
# a lazy model that always predicts "not fraud"
correct = total - fraud          # it's right on all 990 legit ones
accuracy = correct / total
recall_fraud = 0 / fraud         # but catches none of the fraud
print(f"accuracy    : {accuracy:.1%}")
print(f"fraud recall: {recall_fraud:.0%}")
print(f"caught 0 of {fraud} frauds — despite 99% accuracy")
```
