## A monitor that watches both layers
Each metric has acceptable bounds; anything outside raises an alert. The first two are operational (system health), the third is model quality — watch which one fires.

```python
def alert(name, value, lo=None, hi=None):
    bad = (lo is not None and value < lo) or (hi is not None and value > hi)
    print(f"{name:12s} = {value:<5} [{'ALERT' if bad else 'ok'}]")

alert("latency_ms", 95,   hi=200)    # system health: under 200ms budget
alert("error_rate", 0.01, hi=0.05)   # system health: under 5% errors
alert("accuracy",   0.78, lo=0.80)   # model quality: must stay above 0.80
```
