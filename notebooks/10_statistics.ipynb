# Lesson 1.10 — Statistics & MLE

```python
import numpy as np
s=np.array([2,4,4,4,5,5,7,9])
print('mean:', s.mean(), ' var:', s.var(), ' std:', s.std())   # 5, 4, 2
```

## Law of Large Numbers on a die (true mean 3.5)

```python
rng=np.random.default_rng(0)
for n in [10,100,10000]:
    print(f'n={n:5d}  sample mean = {rng.integers(1,7,size=n).mean():.3f}')
```

## Maximum likelihood for a coin (7 heads in 10)

```python
from math import comb
heads,n=7,10; ps=np.linspace(0,1,101)
like=comb(n,heads)*ps**heads*(1-ps)**(n-heads)
print('MLE of p:', round(ps[np.argmax(like)],2))   # 0.70
```

## Central Limit Theorem: averages of a (flat) population become bell-shaped

```python
rng=np.random.default_rng(1)
means=[rng.random(30).mean() for _ in range(5000)]   # each = mean of 30 uniforms
means=np.array(means)
print('mean of sample-means:', round(means.mean(),3), ' std:', round(means.std(),3))
try:
    import matplotlib.pyplot as plt
    plt.hist(means, bins=30); plt.title('distribution of sample means (looks normal)'); plt.show()
except Exception as e:
    print('plot skipped:', e)
```
