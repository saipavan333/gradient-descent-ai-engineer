## Split a real dataset the right way (stratified, before any preprocessing)

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

X, y = load_breast_cancer(return_X_y=True)
print("examples:", X.shape[0], " features:", X.shape[1], " positives:", int(y.sum()))

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0, stratify=y)
print("train:", X_train.shape[0], " test:", X_test.shape[0])
print(f"benign fraction - train: {y_train.mean():.3f}, test: {y_test.mean():.3f}")
```

Stratification keeps the class balance nearly identical in both halves. The test set now stays untouched until the very end.
