## Promotion and instant rollback
`promote` archives the previous production model first, so one model is live at a time. The rollback at the end restores the known-good version with no retraining.

```python
registry = {}
def register(version): registry[version] = "staging"
def promote(version):                          # one model is 'production' at a time
    for v in registry:
        if registry[v] == "production": registry[v] = "archived"
    registry[version] = "production"

register("v1"); promote("v1")
register("v2"); promote("v2")
print("after v2 promote:", registry)
promote("v1")                                  # rollback to the known-good v1
print("after rollback  :", registry)
```
