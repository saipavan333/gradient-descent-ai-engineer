## A minimal word-level tokenizer: encode and decode

```python
text = "the cat sat on the mat"
words = text.split()
vocab = {w: i for i, w in enumerate(sorted(set(words)))}
print("vocab:", vocab)
ids = [vocab[w] for w in words]
print("token ids:", ids)
inv = {i: w for w, i in vocab.items()}
print("decoded:", " ".join(inv[i] for i in ids))
```
