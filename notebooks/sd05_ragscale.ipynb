## The token cost of retrieved context
Retrieved chunks become LLM input tokens; k (chunks retrieved) sets the per-query cost and, at scale, the daily bill.

```python
system_tokens, query_tokens, chunk_tokens, k = 100, 50, 200, 5
input_tokens = system_tokens + query_tokens + k * chunk_tokens
cost_per_1k = 0.003
per_query = input_tokens / 1000 * cost_per_1k
print(f"tokens sent to LLM: {input_tokens}")
print(f"cost per query    : ${per_query:.5f}")
qpd = 1_000_000
print(f"daily cost (1M q) : ${per_query * qpd:,.0f}")
```
