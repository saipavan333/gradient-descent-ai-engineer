## The seed: making 'random' repeatable
Fix the seed and anything 'random' comes out identically every run — the most-forgotten of the five reproducibility determinants.

```python
import numpy as np
def run():
    np.random.seed(42)              # fix the seed before any randomness
    return np.random.rand(3).round(4).tolist()
print("run 1:", run())
print("run 2:", run())
print("identical:", run() == run())
```
