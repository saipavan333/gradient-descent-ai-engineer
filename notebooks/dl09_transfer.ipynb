## Train ONLY a head on frozen features

```python
import numpy as np
rng = np.random.default_rng(0); n, d, k = 300, 5, 32
W_features = rng.normal(0,1,(d,k))           # frozen extractor
def features(X): return np.tanh(X @ W_features)
X_train = rng.normal(0,1,(n,d)); true_head = rng.normal(0,1,(k,1))
y_train = features(X_train) @ true_head + rng.normal(0,0.1,(n,1))
F = features(X_train); head, *_ = np.linalg.lstsq(F, y_train, rcond=None)
print("train MSE:", round(float(((F @ head - y_train)**2).mean()),4))
X_test = rng.normal(0,1,(80,d)); y_test = features(X_test) @ true_head + rng.normal(0,0.1,(80,1))
print("test  MSE:", round(float(((features(X_test) @ head - y_test)**2).mean()),4))
```
