# Lesson 2.6 — Plotting with Matplotlib

```python
import numpy as np
try:
    import matplotlib; matplotlib.use("Agg")
    import matplotlib.pyplot as plt
    rng=np.random.default_rng(0)
    size=rng.normal(1500,400,200).clip(500); price=size*0.2+rng.normal(0,30,200)+50
    fig,ax=plt.subplots(figsize=(5,3)); ax.hist(size,bins=20,color="#6366f1",edgecolor="white")
    ax.set_xlabel("size (sq ft)"); ax.set_ylabel("count"); ax.set_title("Distribution of sizes")
    fig.savefig("/tmp/hist.png"); print("histogram saved")
    fig,ax=plt.subplots(figsize=(5,3)); ax.scatter(size,price,s=12,alpha=0.6,color="#10b981")
    ax.set_xlabel("size"); ax.set_ylabel("price"); ax.set_title("Price vs size")
    fig.savefig("/tmp/scatter.png"); print("scatter saved")
except ImportError:
    print("matplotlib not installed — skip")
```
