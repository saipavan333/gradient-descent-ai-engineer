## Detecting drift and triggering a retrain
Compare a live feature's average to the training baseline; if it has moved too far, retrain. The training mean was 50.

```python
train_mean = 50.0
def drift_score(live):
    return round(abs(sum(live)/len(live) - train_mean) / train_mean, 3)

for label, vals in [("week 1", [48,51,49,52,50]), ("week 8", [70,72,68,71,69])]:
    d = drift_score(vals)
    print(f"{label}: drift={d}  ->", "RETRAIN" if d > 0.10 else "ok")
```
