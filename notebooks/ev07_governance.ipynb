## Redacting PII before it reaches the model
Detect and replace personal information (emails, SSNs) with placeholders — data minimization in practice. (Regex is brittle; real systems use NER.)

```python
import re
def redact(text):
    text = re.sub(r'[\w.]+@[\w.]+\.\w+', '[EMAIL]', text)      # emails
    text = re.sub(r'\b\d{3}-\d{2}-\d{4}\b', '[SSN]', text)    # US SSNs
    return text

print(redact("Email alex@example.com or reference SSN 123-45-6789 on file."))
```
