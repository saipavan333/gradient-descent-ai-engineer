## Train a tiny network to learn XOR (manual backprop + SGD)

```python
import numpy as np
rng = np.random.default_rng(1)
X = np.array([[0,0],[0,1],[1,0],[1,1]], float)
Y = np.array([[0],[1],[1],[0]], float)
W1 = rng.normal(0,1,(2,8)); b1 = np.zeros((1,8))
W2 = rng.normal(0,1,(8,1)); b2 = np.zeros((1,1))
sig = lambda z: 1/(1+np.exp(-z)); lr = 0.3
for epoch in range(4000):
    A1 = np.tanh(X @ W1 + b1); P = sig(A1 @ W2 + b2)
    if epoch == 0: start = np.mean((P-Y)**2)
    dZ2 = 2*(P-Y)*P*(1-P)/4
    dW2 = A1.T @ dZ2; db2 = dZ2.sum(0, keepdims=True)
    dZ1 = (dZ2 @ W2.T) * (1 - A1**2)
    dW1 = X.T @ dZ1; db1 = dZ1.sum(0, keepdims=True)
    W1-=lr*dW1; b1-=lr*db1; W2-=lr*dW2; b2-=lr*db2
print("start MSE:", round(start,4), " final MSE:", round(np.mean((P-Y)**2),5))
print("predictions:", P.round(2).ravel(), " targets:", Y.ravel())
```
