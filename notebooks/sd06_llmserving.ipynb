## Sizing the GPU fleet
LLM capacity works in tokens/sec: demand = QPS x tokens/response; GPUs = demand / per-GPU throughput.

```python
import math
qps, tokens_per_response, gpu_tokens_per_sec = 100, 500, 2500
demand = qps * tokens_per_response          # total tokens/sec the platform must generate
gpus = math.ceil(demand / gpu_tokens_per_sec)
print(f"tokens/sec demand: {demand:,}")
print(f"GPUs needed      : {gpus}")
```
