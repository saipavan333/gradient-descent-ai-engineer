## A minimal experiment tracker
Log each run's params and metrics, then find the best run by its metric — the comparison that tracking makes trivial.

```python
runs = []
def log_run(name, params, metrics):
    runs.append({"name": name, "params": params, "metrics": metrics})

log_run("run1", {"lr": 0.10, "layers": 2}, {"val_acc": 0.71})
log_run("run2", {"lr": 0.01, "layers": 4}, {"val_acc": 0.86})
log_run("run3", {"lr": 0.01, "layers": 8}, {"val_acc": 0.83})

best = max(runs, key=lambda r: r["metrics"]["val_acc"])
print("logged", len(runs), "runs")
print("best:", best["name"], best["params"], "->", best["metrics"]["val_acc"])
```
