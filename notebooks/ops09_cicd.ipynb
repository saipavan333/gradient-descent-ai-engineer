## A deployment gate with an ML-specific check
A change only deploys if the tests pass AND the new model beats the production baseline — the gate that makes automated retraining safe.

```python
def ci_pipeline(tests_pass, new_acc, baseline_acc):
    if not tests_pass:          return "BLOCKED: tests failed"
    if new_acc < baseline_acc:  return "BLOCKED: model worse than baseline"
    return "DEPLOY: all gates passed"

print(ci_pipeline(True,  0.88, 0.86))   # tests pass, beats baseline
print(ci_pipeline(False, 0.88, 0.86))   # tests fail
print(ci_pipeline(True,  0.84, 0.86))   # worse than baseline
```
