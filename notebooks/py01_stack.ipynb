# Lesson 2.1 — The scientific Python stack

```python
import numpy as np, pandas as pd
rng = np.random.default_rng(42)
heights = rng.normal(170, 8, size=5).round(1)
weights = rng.normal(70, 10, size=5).round(1)
people = pd.DataFrame({"height_cm": heights, "weight_kg": weights})
print(people)
print("\nsummary:\n", people.describe().round(1))
```
