## Initialization scale controls activation spread across depth

```python
import numpy as np
rng = np.random.default_rng(0); n = 256
for scale in [0.02, round(float(np.sqrt(1/n)),3), 0.2]:
    a = rng.normal(0,1,n); stds = []
    for layer in range(6):
        W = rng.normal(0, scale, (n,n)); a = W @ a; stds.append(round(float(a.std()),3))
    print(f"weight scale {scale}: activation std by layer {stds}")
```
