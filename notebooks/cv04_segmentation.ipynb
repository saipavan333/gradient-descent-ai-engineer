## Per-pixel accuracy: compare a predicted mask to the truth


```python
import numpy as np
true = np.array([[0,0,1],[0,1,1],[2,2,1]])      # each cell = the class of that pixel
pred = np.array([[0,0,1],[0,1,1],[2,2,2]])      # the model's per-pixel prediction
print("per-pixel accuracy:", round((true == pred).mean(), 3))
print("pixels correct:", int((true == pred).sum()), "/", true.size)
```

## Explore: why per-class IoU beats raw accuracy under imbalance


```python
# IoU for one class = overlap pixels / union pixels (for that class).
def class_iou(t, p, cls):
    inter = ((t==cls) & (p==cls)).sum()
    union = ((t==cls) | (p==cls)).sum()
    return inter/union if union else float("nan")
for c in [0,1,2]:
    print(f"class {c}: IoU = {class_iou(true,pred,c):.3f}")
# Class 2 is penalized (predicted where it should not be) even though
# overall pixel accuracy stays high -> per-class IoU exposes the error.
```
