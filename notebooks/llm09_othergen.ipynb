## The VAE reparameterization trick: z = mu + sigma * eps

```python
import numpy as np
rng = np.random.default_rng(0)
mu, sigma = np.array([0.5, -0.2]), np.array([0.3, 0.4])
samples = np.array([mu + sigma * rng.standard_normal(2) for _ in range(5)])
print("encoder gives  mu =", mu, " sigma =", sigma)
print("5 sampled latents z:\n", samples.round(2))
```
