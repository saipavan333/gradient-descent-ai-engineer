# Lesson 1.6 — Derivatives & gradients

```python
import numpy as np
def f(x,y): return x**2 + y**2
h=1e-5
dfdx=(f(1+h,2)-f(1-h,2))/(2*h); dfdy=(f(1,2+h)-f(1,2-h))/(2*h)
print('numerical gradient at (1,2):', round(dfdx,3), round(dfdy,3))  # 2,4
print('analytic gradient (2x,2y):', 2*1, 2*2)
```

## A downhill step reduces the function

```python
g=np.array([2.,4.]); p=np.array([1.,2.]); lr=0.1
print('f before:', f(*p), ' f after:', round(f(*(p-lr*g)),3))   # 5 -> 3.2
```

## Autograd gets exact gradients automatically (optional)

```python
try:
    import torch
    x=torch.tensor(1.,requires_grad=True); y=torch.tensor(2.,requires_grad=True)
    (x**2+y**2).backward(); print('torch grads:', x.grad.item(), y.grad.item())
except ImportError:
    print('PyTorch not installed — skip')
```
