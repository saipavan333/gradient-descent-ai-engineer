## Sinusoidal positional encodings

```python
import numpy as np
def positional_encoding(seq_len, d):
    pos = np.arange(seq_len)[:, None]; i = np.arange(d)[None, :]
    angle = pos / (10000 ** (2*(i//2)/d))
    return np.where(i % 2 == 0, np.sin(angle), np.cos(angle))
P = positional_encoding(4, 6)
print("positional encodings (4 positions, d=6):\n", P.round(2))
print("position 0 identical to position 1?", np.allclose(P[0], P[1]))
```
