## Serve predictions while logging every request
The core pattern: every request flows through a path that both serves the prediction and records it, so the service is observable.

```python
log = []
def predict(x): return x * 2                    # the "model"
def serve(x):
    y = predict(x)
    log.append({"input": x, "output": y})        # monitoring: log every request
    return y
for x in [3, 7, 10]:
    print(f"serve({x}) -> {serve(x)}")
print(f"logged {len(log)} requests for monitoring")
print(f"avg output: {sum(r['output'] for r in log)/len(log):.1f}")
```
