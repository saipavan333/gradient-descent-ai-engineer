# Lesson 1.2 — Matrices & matrix multiplication (runnable companion)

```python
import numpy as np
# A dataset is a matrix: rows = examples, columns = features
X = np.array([[1800, 3, 12],
              [ 950, 2, 40],
              [2400, 4,  5]])
print("shape:", X.shape, " (3 houses, 3 features)")
```

## Matrix x vector: each output is one row 'mixed with' the input

```python
A = np.array([[1,0,2],[0,3,1]])
x = np.array([4,1,2])
print("A x =", A @ x)        # [8, 5]; a 2x3 times a 3-vector -> a 2-vector
```

## Matrix x matrix: entry (i,j) = row i . column j

```python
A = np.array([[1,2,0],[0,1,3]])      # 2x3
B = np.array([[2,1],[1,0],[4,2]])    # 3x2
print("A @ B =\n", A @ B)            # [[4,1],[13,6]]; 2x2
```

## The shape rule, demonstrated (this catches most real bugs)

```python
def try_mul(s1, s2):
    try:
        r = (np.ones(s1) @ np.ones(s2)).shape
        print(f"{s1} @ {s2} -> {r}  OK")
    except ValueError as e:
        print(f"{s1} @ {s2} -> ILLEGAL ({e})")
try_mul((5,3),(3,8))   # inner 3==3 -> 5x8
try_mul((2,3),(2,3))   # inner 3 != 2 -> illegal
```

## Batch prediction = one matrix multiply (a linear model)

```python
w = np.array([0.18, 15, -1.2])       # untrained example weights
print("predicted prices ($k):", X @ w)   # [354.6, 153., 486.]
```

## A linear layer by hand: y = f(Wx + b)

```python
W = np.array([[0.18, 15, -1.2],
              [0.05,  8, -0.3]])      # 2 outputs, 3 inputs
b = np.array([20.0, 5.0])
relu = lambda z: np.maximum(0, z)     # a simple activation
Y = relu(X @ W.T + b)
print("layer output shape:", Y.shape) # (3, 2)
print(Y)
```
