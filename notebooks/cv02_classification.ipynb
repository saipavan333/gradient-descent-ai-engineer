## One convolution by hand: an edge detector over a square


```python
import numpy as np
img = np.array([[0,0,0,0,0],
                [0,9,9,9,0],
                [0,9,9,9,0],
                [0,9,9,9,0],
                [0,0,0,0,0]], float)
kernel = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]], float)   # edge detector
H,W,k = *img.shape, 3
out = np.zeros((H-k+1, W-k+1))
for i in range(H-k+1):
    for j in range(W-k+1):
        out[i,j] = (img[i:i+k, j:j+k] * kernel).sum()          # multiply + sum
print("feature map:\n", out.astype(int))
```

## Explore: the filter fires on the border, stays silent in flat interior


```python
# The center (interior of the square) is flat -> response 0.
# The corners/edges (transition light<->dark) -> large response.
print("center response:", int(out[1,1]), "(flat interior -> no edge)")
print("corner response:", int(out[0,0]), "(a border -> strong edge)")
# The same 3x3 filter was reused at every position: weight sharing.
```
