## One turn of the lifecycle loop
Five stages run in order; the final `monitor` stage decides whether the deployed model is healthy or needs retraining — the feedback that closes the loop.

```python
def collect():        return "1,000 labeled rows"
def train(data):      return {"name": "fraud-v1", "data": data}
def evaluate(model):  return 0.86                      # validation accuracy
def deploy(model):    return f"{model['name']} serving live"
def monitor(acc):     return "healthy" if acc >= 0.80 else "RETRAIN"

data   = collect()
model  = train(data)
acc    = evaluate(model)
status = deploy(model)
print("1 data    :", data)
print("2 train   :", model["name"])
print("3 evaluate:", acc)
print("4 deploy  :", status)
print("5 monitor :", monitor(acc))
```
