## Cosine similarity and the king - man + woman analogy

```python
import numpy as np
emb = {"king":np.array([0.9,0.1,0.8]),"queen":np.array([0.9,0.9,0.8]),
       "man":np.array([0.1,0.1,0.2]),"woman":np.array([0.1,0.9,0.2]),
       "apple":np.array([0.2,0.5,-0.7])}
cos = lambda a,b: a@b/(np.linalg.norm(a)*np.linalg.norm(b))
print("similarity(king, queen):", round(cos(emb["king"],emb["queen"]),3))
print("similarity(king, apple):", round(cos(emb["king"],emb["apple"]),3))
analogy = emb["king"]-emb["man"]+emb["woman"]
best = max((w for w in emb if w not in ["king","man","woman"]), key=lambda w: cos(analogy,emb[w]))
print("king - man + woman  is closest to:", best)
```
